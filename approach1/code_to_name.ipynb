```python
from bs4 import BeautifulSoup
import requests
import warnings
import pandas as pd
from datetime import datetime
pd.options.mode.chained_assignment = None
```

```python
df = pd.read_csv("stations.csv")
df
```

```python
station_codes = df['S_No'].to_list()
i, k = 7575, 0
data = []
while i < len(station_codes):
     try:
          link = f"https://indiarailinfo.com/departures/{station_codes[i]}?date=undefined&stptype=1&"
          headers = {
               "user-agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.41 Safari/537.36"
          }

          req = requests.get(link, headers=headers)
          soup = BeautifulSoup(req.content, "lxml")
          get_top_content = soup.find("div", class_="topcapsule").h1.contents[0]
          x, y = get_top_content.find("from")+4, get_top_content.find(")")
          pfs, s_code = 0, station_codes[i]
          #print(og_station_name)
          if y != -1:
               og_station_name = get_top_content[x:y+1]
               if "PFs" in og_station_name:
                    pfs = og_station_name[og_station_name.find("(")+1:-5]
                    station_name = og_station_name[og_station_name.find("/")+1:og_station_name.find("(")-1]
               else:
                    station_name = og_station_name[og_station_name.find("/")+1:y+1]
          else:
               og_station_name = get_top_content[x:]
               station_name = og_station_name[og_station_name.find("/")+1:]
               print(station_name)

          # txt = f"{s_code}, {station_name}, {pfs}"
          # print(txt)
          with open("stations2.txt", "a") as f:
               print(f"{s_code}, {station_name}, {pfs}", file=f)
          i+=1
          k+=1
     
     except Exception as e:
          print(e)
          print(f"Program ended at {i} sequence")
          break
     # print(i)
```
